```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df_cg = pd.read_csv("../data/genotypes/chr4//genotypes_region.csv",index_col=0)
df_cg = df_cg.iloc[:, :200]
df_cg["label"] = 0
df_cg.describe()
```

```python
df_chr4 = pd.read_csv("../data/genotypes/chr4/genotypes_region-chr4.csv",index_col=0)
df_chr4 = df_chr4.iloc[:, :200]
df_chr4["label"] = 1
df_chr4.describe()
```

```python
df_cg_encoded = pd.get_dummies(df_cg, columns=list(df_cg.columns[:-1]), drop_first=True, dtype=int)
df_chr4_encoded = pd.get_dummies(df_chr4, columns=list(df_chr4.columns[:-1]), drop_first=True, dtype=int)
df_concat = pd.concat([df_cg_encoded, df_chr4_encoded])
df_concat = df_concat.fillna(0)
df_concat_shuffled = df_concat.iloc[np.random.permutation(len(df_concat))]
df_concat_shuffled.to_csv("../data/genotypes/chr4/dataset.csv")
```

```python
y = df_concat_shuffled.iloc[:,0].to_numpy()
X = df_concat_shuffled.iloc[:,1:].to_numpy()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
clf = LogisticRegression(random_state=0).fit(X_train,y_train)
```

```python
clf.score(X_test, y_test)
```
